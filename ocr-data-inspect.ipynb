```python
import pandas as pd
import numpy as np
from glob import glob
import os
import matplotlib.pyplot as plt
import cv2

# os.environ["CUDA_VISIBLE_DEVICES"]="1"

ROOT_DIR = os.getcwd()
ANNOTATIONS_FILE = os.path.join(ROOT_DIR, 'data/class_annotations.csv')
CLASSES_FILE = os.path.join(ROOT_DIR, 'data/class_mapping.csv')

df = pd.read_csv(ANNOTATIONS_FILE, names = ["name", "x1", "y1", "x2", "y2", "x3", "y3", "x4", "y4", "classname"])
df.name.unique()

# mapping = pd.read_csv(CLASSES_FILE, names=['name', 'id'])
# mapping
 
series = df.loc[df['name'] == 'image/imageresized_0.jpg']
print(series.shape)

series = df.loc[df['name'] == 'image/imageresized_1.jpg']
print(series.shape)

def draw_polygon(image, box, color):
    """ Draws a polygon on an image with a given color.

    # Arguments
        image     : The image to draw on.
        box       : A list of 4 elements (x1, y1, x2, y2).
        color     : The color of the box.
        thickness : The thickness of the lines to draw a box with.
    """
    b = np.array(box).astype(int)
    pts = np.array([ [b[0],b[1]], [b[2],b[3]], [b[4],b[5]], [b[6],b[7]] ], np.int32)
    pts = pts.reshape((-1,1,2))
    image = cv2.fillPoly(image, [pts], 255)
    return image

class_mapping = {
            "product_id": 1,
            "order_id": 2,
        }

instance_masks = []
class_ids = []

# for seri in series:
mask = np.zeros((1024, 1024, 3), dtype=np.uint8)
# plt.figure(figsize=(15, 15))
# plt.imshow(mask)
# plt.show()

for idx, row in series.iterrows():
#     print(idx)
#     print(idx)
#     print(row[['x1', 'y1']].reshape(-1))
#     print(row.classname)
    class_id = class_mapping[row.classname]
    print(class_id)
    if class_id:
        box = row[['x1', 'y1', 'x2', 'y2', 'x3', 'y3', 'x4', 'y4']].values.reshape(-1)
        print(box)
        m = draw_polygon(mask.copy(), box, 1)
        plt.figure(figsize=(15, 15))
        plt.imshow(m)
        plt.show()
        instance_masks.append(m)
        class_ids.append(class_id)
        
```
